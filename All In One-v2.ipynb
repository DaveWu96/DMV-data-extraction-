```python
text = "15002139 Sustained/Disclose"

def IS_No(text):
    nlp_vocab = nlp.vocab
    matcher1 = Matcher(nlp.vocab)

    #Pattern 1
    Nopattern = [
        {'POS': 'NUM'}   
    ]


    matcher1.add("matcher", [Nopattern]) #enable for IS No. pattern
    print("----IS_No----")
#     for i in df.index:

#         text = df['IS No.'][i]     #enable for IS No. pattern
#         print(text)
    doc = nlp(text)
    matches1 = matcher1(doc)

    #  span2 so it doesn't include "Case:" ---- CASE        
    for match_id, start, end in matches1:
        string_id = nlp.vocab.strings[match_id]  
        span1 = doc[start: end]
        count = len(str(span1))
        split = str(span1).split()
        split2 = split[0][0:2].split()
        split3 = split[0][2:count].split()
        holder = [split2[0], split3[0]]
        final = "-".join(holder)

    print(final)
    return final
        
IS_No(text)
```

```python
text = "Incident: 9/4/2015 Disposition: 11/30/2015 DRB: random"
def Dates(text):

    matcher9 = Matcher(nlp.vocab)
    matcher10 = Matcher(nlp.vocab)
    matcher11 = Matcher(nlp.vocab)



    ##Dates
    #Incident_Date
    IDpattern = [
        {'LOWER' : {'IN': ['incident'] } },
        {'POS': 'PUNCT'},
        {'OP': '*', 'LOWER' : {'NOT_IN': ['disposition'] } }
    ]

    DIDpattern = [
        {'LOWER' : {'IN': ['disposition'] } },
        {'POS': 'PUNCT'},
        {'OP': '*', 'LOWER' : {'NOT_IN': ['drb'] }}
    ]

    DRDpattern = [
        {'LOWER' : {'IN': ['drb'] } },
        {'POS': 'PUNCT'},
        {'OP': '*'}
    ]




    matcher9.add("matcher9", [IDpattern])
    matcher10.add("matcher10", [DIDpattern])
    matcher11.add("matcher11", [DRDpattern])
    print("----Dates----")
#     for i in df.index:
#     #text = df['IS No.'][i]     #enable for IS No. pattern
    
#         text = df['Dates'][i]
    doc = nlp(text)
    matches9 = matcher9(doc)
    matches10 = matcher10(doc)
    matches11 = matcher11(doc)
    
    #  span2 so it doesn't include "Case:" ---- CASE        
    for match_id, start, end in matches9:
        string_id = nlp.vocab.strings[match_id]  
        span9 = doc[start: end]
        span9 = span9[2:len(span9)]

    print("Incident: "+ span9.text)


    for match_id, start, end in matches10:
        string_id = nlp.vocab.strings[match_id]  
        span10 = doc[start: end]
        span10 = span10[2:len(span10)]

    print("Disposition: " + span10.text)


    for match_id, start, end in matches11:
        string_id = nlp.vocab.strings[match_id]  
        span11 = doc[start: end]
        span11 = span11[2:len(span11)]

    print("DRB: " + span11.text)
    return span9,span10,span11


Dates(text)
```

```python
text = "Case: Sustained Member Disposition: Misconduct Allegations - - Orders/Directives Violation : Sustained IAD Discipline: Official Reprimand DRB Disposition: Something is here to hold a place DRB Discipline: Something is here to hold a place"
def Disposition(text):
    matcher2 = Matcher(nlp.vocab)
    matcher3 = Matcher(nlp.vocab)
    matcher4 = Matcher(nlp.vocab)
    matcher5 = Matcher(nlp.vocab)
    matcher6 = Matcher(nlp.vocab)



    # Case_Outcome
    Dispattern = [
        {'LOWER' : {'IN': ['case'] } },
        {'POS': 'PUNCT'},
        {'OP': '*', 'LOWER' : {'NOT_IN': ['member'] } }
    ]


    #Member_Disposition
    Mempattern = [
        {'LOWER' : {'IN': ['member'] } },
        {'LOWER' : {'IN': ['disposition'] } },
        {'POS': 'PUNCT'},
        {'OP': '*', 'LOWER' : {'NOT_IN': ['iad'] } }
    ]

    #IAD_Discipline
    IADpattern = [
        {'LOWER' : {'IN': ['iad'] } },
        {'LOWER' : {'IN': ['discipline'] } },
        {'POS': 'PUNCT'},
        {'OP': '*', 'LOWER' : {'NOT_IN': ['drb'] } }
    ]

    #DRB_Disposition
    DRBpattern = [
        {'LOWER' : {'IN': ['drb'] } },
        {'LOWER' : {'IN': ['disposition'] } },
        {'POS': 'PUNCT'},
        {'OP': '*', 'LOWER' : {'NOT_IN': ['drb'] } }
    ]

    #DRB_Discipline
    DRB2pattern = [
        {'LOWER' : {'IN': ['drb'] } },
        {'LOWER' : {'IN': ['discipline'] } },
        {'POS': 'PUNCT'},
        {'OP': '*'}
    ]





    matcher2.add("matcher2", [Dispattern])
    matcher3.add("matcher3", [Mempattern])
    matcher4.add("matcher4", [IADpattern])
    matcher5.add("matcher5", [DRBpattern])
    matcher6.add("matcher6", [DRB2pattern])
    
    print("----Disposition----")
#     for i in df.index:
    
#         text = df['Disposition'][i]
#         print(text)
    doc = nlp(text)
    matches2 = matcher2(doc)
    matches3 = matcher3(doc)
    matches4 = matcher4(doc)
    matches5 = matcher5(doc)
    matches6 = matcher6(doc)

#  span2 so it doesn't include "Case:" ---- CASE        
    for match_id, start, end in matches2:
        string_id = nlp.vocab.strings[match_id]  
        span2 = doc[start: end]
        span2 = span2[2:len(span2)]
    print("Case: " + span2.text)

# span3 so it doesn't include "Member Disposition:" ------- Member Disposition        
    for match_id, start, end in matches3:
        string_id = nlp.vocab.strings[match_id]  
        span3 = doc[start: end]
        span3 = span3[3:len(span3)]
    print("Member Disposition: " + span3.text)

# span4 for IAD    
    for match_id, start, end in matches4:
        string_id = nlp.vocab.strings[match_id]  
        span4 = doc[start: end]
        span4 = span4[3:len(span4)]
    print("IAD Discipline: " + span4.text)

    for match_id, start, end in matches5:
        string_id = nlp.vocab.strings[match_id]  
        span5 = doc[start: end]
        span5 = span5[3:len(span5)]
    print("DRB Disposition: " + span5.text)

    for match_id, start, end in matches6:
        string_id = nlp.vocab.strings[match_id]  
        span6 = doc[start: end]
        span6 = span6[3:len(span6)]
    print("DRB Discipline: " + span6.text)
    return span2,span3,span4,span5,span6

Disposition(text)
```

```python
text = "Synopsis: On 9-4-15 members arrested Defendant ||||||||||| for AWIK that took place on 8-30-15 in the 1200 block of Crittenden Street NW.  A firearm was found in the prisoner's compartment of the transport vehicle that took Defendant ||||||||||| to Fourth District for processing.Investigative Notes: received IAB 11/30/15 - SUSTIANED for all members involved - Official Reprimands were issued to Banks, Symanski, and Thompson. Adverse Action for Alacron and Molinari. -AR"
def Case(text):

    matcher7 = Matcher(nlp.vocab)
    matcher8 = Matcher(nlp.vocab)



    ##Case Description
    #Synopsis
    Synpattern = [
        {'LOWER' : {'IN': ['synopsis'] } },
        {'POS': 'PUNCT'},
        {'OP': '*', 'ORTH' : {'NOT_IN': ['Investigative'] } }
    ]

    Invpattern = [
        {'LOWER' : {'IN': ['investigative'] } },
        {'LOWER' : {'IN': ['notes'] } },
        {'POS': 'PUNCT'},
        {'OP': '*'}
    ]





    matcher7.add("matcher7", [Synpattern])
    matcher8.add("matcher8", [Invpattern])
    print("----Case Description----")
    # for i in df.index:
    #text = df['IS No.'][i]     #enable for IS No. pattern
    
    # text = df['Case Description'][i]
    doc = nlp(text)
    matches7 = matcher7(doc)
    matches8 = matcher8(doc)

#  span2 so it doesn't include "Case:" ---- CASE        
    for match_id, start, end in matches7:
        string_id = nlp.vocab.strings[match_id]  
        span7 = doc[start: end]
        span7 = span7[2:len(span7)]
    print("Synopsis: " + span7.text)


    for match_id, start, end in matches8:
        string_id = nlp.vocab.strings[match_id]  
        span8 = doc[start: end]
        span8 = span8[3:len(span8)]
    print("Investigative Notes: " + span8.text)
    return span7, span8

Case(text)
```

```python
text = "MPD sustained a finding for a violation of General Order 120.21, Table A, Number 16, for failure to conduct a field search of a prisoner. For this sustained finding, officer Alarcon was disciplined with the forfeiture of 40 hours of annual leave. IS-Related Documents: IS15002139.pdf IS15002139 - Symanski.pdf IS15002139 - Banks.pdf IS15002139 - Alarcon.pdf IS15002139 - Molinari.pdf"

def Additional(text):
    matcher13 = Matcher(nlp.vocab)
 
    IS_related_pattern = [
        {'ORTH': 'IS'},
        {'OP': '*'}

    ]


    matcher13.add("matcher13", [IS_related_pattern])
    print("----Additonal Information----")
#     for i in df.index:

#         text = df['Additional Information'][i]
#         print(text)
    doc = nlp(text)
    matches13 = matcher13(doc)

    for match_id, start, end in matches13:
        string_id = nlp.vocab.strings[match_id]  
        span13 = doc[start: end]

    text = text.split()
    span13 = str(span13)
    span13 = span13.split()
    count = len(text) - len(span13)
    if count < 0:
        span14 = ''
    else:
        span14 = text[0: count]
        span14 = " ".join(span14)
    print(span14)
    return span14
    
Additional(text)
```

```python
def Header(doc):
    
    matcher14 = Matcher(nlp.vocab)

    Headpattern = [
        {'ORTH': 'CAD'},
        {'OP': '*'},
        {'POS': 'NUM'}   
    ]

    matcher14.add("matcher14", [Headpattern]) 
    
    header = []
    for token in doc[0:1000]:
        header.append(str(token))
        for elem in (header):
            if elem == '\n' or elem =='\n\n':
                header.remove(elem)
        text = ' '.join(header)
    print(text)   

    doc = nlp(text)
    matches14 = matcher14(doc)


    for match_id, start, end in matches14:
        string_id = nlp.vocab.strings[match_id]  
        span15 = doc[start: end]
        span15 = span15[0:11]
        span15 = str(span15)
    print("ANSWER: " + span15)
    return span15

Header(doc)
```
